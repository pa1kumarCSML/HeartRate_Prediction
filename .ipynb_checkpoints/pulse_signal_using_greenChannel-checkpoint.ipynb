## Step-by-Step Explanation of the Code

### 1. Generate Simulated Heart Rate Data

```python
heart_rate = np.sin(np.linspace(0, 2 * np.pi, len(green_avg))) * 10 + 60
```

This line generates simulated heart rate data based on a sine wave:
- `np.linspace(0, 2 * np.pi, len(green_avg))` generates an array of evenly spaced values from \(0\) to \(2\pi\), with the length equal to `len(green_avg)`.
- `np.sin(...)` applies the sine function to these values, creating a sine wave.
- `* 10` scales the sine wave to have a range of \(\pm 10\).
- `+ 60` shifts the sine wave so that it oscillates around 60, simulating heart rate values around 60 beats per minute (bpm).

### 2. Fit a Linear Regression Model

```python
model = LinearRegression()
model.fit(green_avg.reshape(-1, 1), heart_rate)
```

- `LinearRegression()` initializes a linear regression model.
- `green_avg.reshape(-1, 1)` reshapes `green_avg` to be a 2D array with one column.
- `model.fit(...)` fits the linear regression model to the data, using `green_avg` as the predictor (independent variable) and `heart_rate` as the response (dependent variable).

### 3. Predict Heart Rate

```python
predicted_heart_rate = model.predict(green_avg.reshape(-1, 1))
```

This line uses the fitted linear regression model to predict heart rates based on the `green_avg` data.

### 4. Bandpass Filter

```python
fs = 30.0
lowcut = 0.75
highcut = 2.5
filtered_green_avg = butter_bandpass_filter(green_avg, lowcut, highcut, fs, order=4)
```

- `fs = 30.0`: Sampling frequency, 30 Hz in this case.
- `lowcut = 0.75`: Lower cutoff frequency of the bandpass filter.
- `highcut = 2.5`: Upper cutoff frequency of the bandpass filter.
- `butter_bandpass_filter(...)` applies a Butterworth bandpass filter to `green_avg`, keeping frequencies between `0.75` and `2.5` Hz, which correspond to heart rates between 45 bpm and 150 bpm. The `order=4` parameter defines the filter's order, affecting its steepness.

### 5. Find Peaks in Filtered Signal

```python
peaks, _ = find_peaks(filtered_green_avg, distance=fs*0.6)
```

- `find_peaks(...)` detects peaks in the `filtered_green_avg` signal.
- `distance=fs*0.6` ensures that detected peaks are at least 0.6 seconds apart, corresponding to a maximum heart rate of 100 bpm.

### 6. Calculate Heart Rate from Peak Intervals

```python
peak_intervals = np.diff(peaks) / fs
heart_rate_estimated = 60.0 / peak_intervals
```

- `np.diff(peaks)`: Calculates the differences between consecutive peak indices.
- `/ fs`: Converts these differences from sample counts to seconds.
- `60.0 / peak_intervals`: Converts intervals in seconds to heart rate in bpm.

### 7. Compute Average Estimated Heart Rate

```python
average_heart_rate = np.mean(heart_rate_estimated)
```

This line calculates the average of the estimated heart rates obtained from the intervals between detected peaks.

---

By following these steps, the code simulates heart rate data, fits a linear regression model, applies a bandpass filter, detects peaks, and computes the estimated heart rate from the peak intervals, finally averaging these estimates to get the final heart rate.

## Selection of Cutoff Frequencies for Bandpass Filter

Selecting the cutoff frequencies for a bandpass filter involves understanding the typical frequency range of the target signal and the specific requirements of the application, especially when processing physiological signals like heart rate from photoplethysmogram (PPG) signals.

### Understanding Heart Rate Frequency Range

Human heart rates typically fall within a specific range, which can be translated into frequency:
- **Resting heart rate:** Generally between 40 and 100 beats per minute (bpm).
- **Active heart rate:** Can go up to 180 bpm or more depending on physical exertion and fitness level.

These rates can be converted to frequency in Hertz (Hz):
- **Lower bound:** 40 bpm = \( \frac{40}{60} \) Hz â‰ˆ 0.67 Hz.
- **Upper bound:** 180 bpm = \( \frac{180}{60} \) Hz = 3 Hz.

### Selection of Cutoff Frequencies

In the code, the selected cutoff frequencies are:
- **Low cutoff frequency (`lowcut`):** 0.75 Hz
- **High cutoff frequency (`highcut`):** 2.5 Hz

These choices are made to capture the typical range of heart rate frequencies while excluding irrelevant frequencies.

#### Reasons for the Specific Choices

1. **Lowcut (0.75 Hz):**
   - Slightly above the lowest possible heart rate (0.67 Hz) to ensure that noise and very low-frequency drifts are filtered out.
   - This helps in focusing on the primary frequency content of the heart rate signal without interference from very slow changes in the signal.

2. **Highcut (2.5 Hz):**
   - Below the upper end of the highest heart rates (3 Hz) to avoid including high-frequency noise or other physiological signals that may not be related to heart rate.
   - Ensures that higher frequency noise or artifacts are removed, which improves the clarity of the heart rate signal.

### Practical Considerations

- **Sampling Frequency (`fs`):** 30 Hz
  - The Nyquist theorem states that the sampling frequency should be at least twice the highest frequency of interest. Here, 30 Hz is much higher than twice the upper cutoff frequency (2.5 Hz), ensuring accurate sampling.
  
- **Order of the Filter (`order=4`):**
  - The order of the filter affects the sharpness of the cutoff. A 4th-order Butterworth filter provides a balance between a sharp cutoff and computational efficiency.

### Summary

The selected cutoff frequencies (0.75 Hz and 2.5 Hz) are chosen to effectively isolate the heart rate signal from noise and other physiological artifacts, while encompassing the typical heart rate range for humans. The exact values may vary slightly depending on the specific application, the nature of the data, and empirical tuning, but the rationale remains the same: to filter out irrelevant frequencies and focus on the heart rate signal.

## Explanation of the Simulated Heart Rate Equation

The equation `heart_rate = np.sin(np.linspace(0, 2 * np.pi, len(green_avg))) * 10 + 60` is used to generate a simulated heart rate signal for demonstration purposes. Let's break down why and how this equation is used:

### Breakdown of the Equation

```python
heart_rate = np.sin(np.linspace(0, 2 * np.pi, len(green_avg))) * 10 + 60
```

1. **`np.linspace(0, 2 * np.pi, len(green_avg))`:**
   - This function generates `len(green_avg)` evenly spaced values between 0 and \(2\pi\).
   - It represents one complete cycle of a sine wave, which corresponds to one heartbeat cycle in this context.

2. **`np.sin(...)`:**
   - The sine function is applied to these values, creating a sine wave.
   - The sine wave oscillates between -1 and 1.

3. **`* 10`:**
   - The sine wave is scaled by a factor of 10.
   - This scaling adjusts the amplitude of the sine wave, so it oscillates between -10 and 10. This represents the variability in the heart rate signal.

4. **`+ 60`:**
   - The entire sine wave is shifted upwards by adding 60.
   - This shifts the range of the sine wave from [-10, 10] to [50, 70].
   - The resulting signal oscillates around 60 beats per minute (bpm), which is a typical resting heart rate for a human.

### Why Use This Equation?

- **Controlled Signal:** A sine wave is a well-understood, smooth, periodic function that is easy to generate and manipulate. Using it allows for a controlled test environment where the characteristics of the signal (such as frequency and amplitude) are known and predictable.
- **Demonstration:** By simulating a heart rate signal, you can demonstrate how your data processing pipeline works (e.g., filtering, peak detection, etc.) without relying on real data, which might have noise and other complexities.
- **Validation:** It provides a simple way to validate that your algorithms for heart rate detection and analysis are working correctly. Since you know the expected output (a heart rate oscillating around 60 bpm), you can easily check if your algorithms produce the correct results.

### Summary

The equation `heart_rate = np.sin(np.linspace(0, 2 * np.pi, len(green_avg))) * 10 + 60` creates a synthetic heart rate signal that oscillates around 60 bpm. This simulated data is useful for testing, demonstrating, and validating heart rate analysis techniques in a controlled manner.

````python
import cv2
import numpy as np
import os
from sklearn.linear_model import LinearRegression
from scipy.signal import find_peaks, butter, filtfilt
import matplotlib.pyplot as plt
import pandas as pd

# Function to perform bandpass filtering
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = filtfilt(b, a, data)
    return y

# Paths
input_folder = '/kaggle/input/ubfc2-video-only'
output_folder = '/kaggle/working/ubfc-2'
os.makedirs(output_folder, exist_ok=True)

# Initialize a list to accumulate results from all videos
all_results = []

# Iterate through all videos in the input folder
for video_file in os.listdir(input_folder):
    if not video_file.endswith('.avi'):
        continue
    
    video_path = os.path.join(input_folder, video_file)
    cap = cv2.VideoCapture(video_path)

    face_regions = {'full_face': [], 'forehead': [], 'right_cheek': [], 'left_cheek': [], 'chin': []}

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
        faces = face_cascade.detectMultiScale(gray, 1.1, 4)
        for (x, y, w, h) in faces:
            face_regions['full_face'].append(frame[y:y+h, x:x+w])
            face_regions['forehead'].append(frame[y:y+h//3, x:x+w])
            face_regions['right_cheek'].append(frame[y+h//3:y+2*h//3, x+w//2:x+w])
            face_regions['left_cheek'].append(frame[y+h//3:y+2*h//3, x:x+w//2])
            face_regions['chin'].append(frame[y+2*h//3:y+h, x:x+w])  # Adding the chin region

    cap.release()

    results = {}
    for region, region_frames in face_regions.items():
        green_avg = []
        for frame in region_frames:
            green_channel = frame[:, :, 1]
            green_avg.append(np.mean(green_channel))
        
        green_avg = np.array(green_avg)
        
        heart_rate = np.sin(np.linspace(0, 2 * np.pi, len(green_avg))) * 10 + 60
        model = LinearRegression()
        model.fit(green_avg.reshape(-1, 1), heart_rate)

        predicted_heart_rate = model.predict(green_avg.reshape(-1, 1))

        fs = 30.0
        lowcut = 0.75
        highcut = 2.5
        filtered_green_avg = butter_bandpass_filter(green_avg, lowcut, highcut, fs, order=4)

        peaks, _ = find_peaks(filtered_green_avg, distance=fs*0.6)

        peak_intervals = np.diff(peaks) / fs
        heart_rate_estimated = 60.0 / peak_intervals

        average_heart_rate = np.mean(heart_rate_estimated)

        results[region] = {
            'green_avg': green_avg,
            'predicted_heart_rate': predicted_heart_rate,
            'filtered_green_avg': filtered_green_avg,
            'peaks': peaks,
            'heart_rate_estimated': heart_rate_estimated,
            'average_heart_rate': average_heart_rate
        }

    # Plotting all regions on the same graph with different colors
    plt.figure(figsize=(12, 6))
    for region, data in results.items():
        plt.plot(data['green_avg'], label=f'{region.capitalize()} Average Green Channel')
        print(data['green_avg'])
    plt.legend()
    plt.title('Average Green Channel Signals for Different Facial Regions')
    plt.xlabel('Frame')
    plt.ylabel('Average Green Channel Signal')
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, f'{os.path.splitext(video_file)[0]}_combined_results.png'))
    plt.close()

    # Plotting individual regions
    plt.figure(figsize=(12, 6))
    for i, (region, data) in enumerate(results.items()):
        plt.subplot(2, 3, i+1)  # Adjust the subplot grid to accommodate the additional chin region
        plt.plot(data['green_avg'], label='Average Green Channel')
        plt.plot(data['filtered_green_avg'], label='Filtered Green Channel')
        plt.scatter(data['peaks'], data['filtered_green_avg'][data['peaks']], color='red', label='Detected Peaks')
        plt.legend()
        plt.title(f'{region.capitalize()} Signal and Detected Peaks')
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, f'{os.path.splitext(video_file)[0]}_individual_results.png'))
    plt.close()

    # Accumulate results for the current video
    for region, data in results.items():
        all_results.append({
            'Video': video_file,
            'Region': region.capitalize(),
            'Average Heart Rate (bpm)': data['average_heart_rate']
        })

# Create a DataFrame from all results and save it as a CSV file
result_table = pd.DataFrame(all_results)
result_table.to_csv(os.path.join(output_folder, 'all_videos_results.csv'), index=False)
````

# Correlation Matrix


A correlation matrix is a table that shows the correlation coefficients between multiple variables. Each cell in the table represents the correlation between two variables, with values ranging from:
- `1`: Perfect positive correlation.
- `-1`: Perfect negative correlation.
- `0`: No correlation.


In the context of the provided code, the correlation matrix is used to analyze the relationship between average green channel signals from different regions of the face (full face, forehead, right cheek, and left cheek) extracted from video frames.

### Steps to Generate the Correlation Matrix

1. **Extracting Face Regions and Computing Green Channel Averages:**
   - For each frame in a video, different regions of the face are extracted: full face, forehead, right cheek, and left cheek.
   - The average intensity of the green channel is computed for each of these regions across all frames.

2. **Storing Signals for Correlation Calculation:**
   - The computed green channel averages for each region are stored in a dictionary.

3. **Calculating the Correlation Matrix:**
   - The correlation coefficients between the green channel signals from different regions are computed.
   - A correlation matrix is formed, where each entry represents the correlation between two regions.

4. **Visualizing the Correlation Matrix:**
   - The correlation matrix is visualized using a heatmap, providing a clear and intuitive representation of the correlations between different facial regions.

### Interpretation

- **High Positive Correlation (near 1):** Indicates that the green channel signals from the two regions are highly similar, moving together in the same direction.
- **High Negative Correlation (near -1):** Indicates that the green channel signals from the two regions are inversely related, moving in opposite directions.
- **Low or Zero Correlation (near 0):** Indicates little to no relationship between the green channel signals from the two regions.

By analyzing the correlation matrix, you can understand how different regions of the face contribute to the overall heart rate signal and how they relate to each other. This can be useful for enhancing signal processing techniques and improving the accuracy of heart rate estimation from facial videos.

````python
import cv2
import numpy as np
import os
from sklearn.linear_model import LinearRegression
from scipy.signal import find_peaks, butter, filtfilt
import matplotlib.pyplot as plt
import pandas as pd

# Function to perform bandpass filtering
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = filtfilt(b, a, data)
    return y

# Paths
input_folder = '/kaggle/input/deepfakeface/DeepFake'
output_folder = '/kaggle/working/test'
os.makedirs(output_folder, exist_ok=True)

# Initialize a dictionary to store signal data for correlation
region_signals = {}

# Iterate through all videos in the input folder
for video_file in os.listdir(input_folder):
    if not video_file.endswith('.avi'):
        continue
    
    video_path = os.path.join(input_folder, video_file)
    cap = cv2.VideoCapture(video_path)

    face_regions = {'full_face': [], 'forehead': [], 'right_cheek': [], 'left_cheek': []}

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
        faces = face_cascade.detectMultiScale(gray, 1.1, 4)
        for (x, y, w, h) in faces:
            face_regions['full_face'].append(frame[y:y+h, x:x+w])
            face_regions['forehead'].append(frame[y:y+h//3, x:x+w])
            face_regions['right_cheek'].append(frame[y+h//3:y+2*h//3, x+w//2:x+w])
            face_regions['left_cheek'].append(frame[y+h//3:y+2*h//3, x:x+w//2])

    cap.release()

    results = {}
    for region, region_frames in face_regions.items():
        green_avg = []
        for frame in region_frames:
            green_channel = frame[:, :, 1]
            green_avg.append(np.mean(green_channel))
        
        green_avg = np.array(green_avg)
        
        # Store region signals for correlation calculation
        region_signals[region] = green_avg

    # Calculate correlations between regions
    correlation_matrix = np.zeros((len(region_signals), len(region_signals)))
    for i, (region1, signal1) in enumerate(region_signals.items()):
        for j, (region2, signal2) in enumerate(region_signals.items()):
            correlation = np.corrcoef(signal1, signal2)[0, 1]
            correlation_matrix[i, j] = correlation

    # Plot correlation matrix
    plt.figure(figsize=(8, 6))
    plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
    plt.colorbar()
    plt.title(f'Correlation Matrix for {os.path.splitext(video_file)[0]}')
    plt.xticks(np.arange(len(region_signals)), list(region_signals.keys()), rotation=45)
    plt.yticks(np.arange(len(region_signals)), list(region_signals.keys()))
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, f'correlation_matrix_{os.path.splitext(video_file)[0]}.png'))
    plt.close()
````

# Z score Normalization to detect anomalies

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore
import pandas as pd
import os

def extract_signal_from_image(image_path):
    # Load the image in grayscale
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    # Check if the image was loaded correctly
    if img is None:
        raise ValueError(f"Image at {image_path} could not be loaded.")
    
    # Assuming the heart signal is represented as variations in pixel intensities
    # along a central horizontal line in the image
    signal = np.mean(img, axis=0)
    
    return signal

def detect_anomalies(signal, threshold=3):
    # Calculate the Z-score for the signal
    z_scores = zscore(signal)
   
    # Identify anomalies where Z-score is beyond the threshold
    anomalies = np.where(np.abs(z_scores) > threshold)
    
    return z_scores, anomalies

# Directory containing heart signal images
image_dir = '/kaggle/working/ubfc-2'  # Replace with your image directory path

# Check if the directory exists
if not os.path.isdir(image_dir):
    raise ValueError(f"Directory {image_dir} does not exist.")

# Prepare a list to store Z-scores and average Z-scores for each image
z_scores_list = []

# Process each image and detect anomalies
for image_name in os.listdir(image_dir):
    image_path = os.path.join(image_dir, image_name)
    
    # Check if the file is an image (optional, based on your files)
    if not image_name.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
        continue
    
    try:
        # Extract the heart signal from the image
        signal = extract_signal_from_image(image_path)
        
        # Detect anomalies in the signal
        z_scores, anomalies = detect_anomalies(signal)
        
        # Calculate the average Z-score
        avg_z_score = np.mean(np.abs(z_scores))
        
        # Store the Z-scores and average Z-score along with the image name
        z_scores_list.append({'image_name': image_name, 'z_scores': z_scores, 'avg_z_score': avg_z_score})
        
        # Plot the signal and highlight anomalies
        plt.figure(figsize=(12, 6))
        plt.plot(signal, label='Heart Signal')
        plt.scatter(anomalies, signal[anomalies], color='red', label='Anomalies')
        plt.title(f'Anomaly Detection in {image_name}')
        plt.xlabel('Index')
        plt.ylabel('Signal Value')
        plt.legend()
        plt.show()
    
    except Exception as e:
        print(f"Error processing {image_name}: {e}")

# Convert the list of dictionaries to a DataFrame
z_scores_df = pd.DataFrame(z_scores_list)

# Since z_scores are numpy arrays, convert them to strings for CSV storage
z_scores_df['z_scores'] = z_scores_df['z_scores'].apply(lambda zs: ','.join(map(str, zs)))

# Save Z-scores and average Z-scores to a CSV file
z_scores_df.to_csv('z_scores.csv', index=False)

print('Z-scores and average Z-scores have been saved to z_scores.csv')
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore
import pandas as pd
import os

def extract_signal_from_image(image_path):
    # Load the image in grayscale
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    # Check if the image was loaded correctly
    if img is None:
        raise ValueError(f"Image at {image_path} could not be loaded.")
    
    # Assuming the heart signal is represented as variations in pixel intensities
    # along a central horizontal line in the image
    signal = np.mean(img, axis=0)
    
    return signal

def detect_anomalies(signal, threshold=3):
    # Calculate the Z-score for the signal
    z_scores = zscore(signal)
   
    # Identify anomalies where Z-score is beyond the threshold
    anomalies = np.where(np.abs(z_scores) > threshold)
    
    return z_scores, anomalies

def frequency_domain_analysis(signal, sampling_rate):
    # Perform Fast Fourier Transform (FFT)
    fft_result = np.fft.fft(signal)
    fft_freq = np.fft.fftfreq(len(signal), d=1/sampling_rate)
    
    return fft_freq, fft_result

# Directory containing heart signal images
image_dir = '/kaggle/working/ubfc-2'  # Replace with your image directory path

# Check if the directory exists
if not os.path.isdir(image_dir):
    raise ValueError(f"Directory {image_dir} does not exist.")

# Prepare a list to store Z-scores, average Z-scores, and frequency domain analysis results for each image
analysis_results = []

# Process each image and detect anomalies
for image_name in os.listdir(image_dir):
    image_path = os.path.join(image_dir, image_name)
    
    # Check if the file is an image (optional, based on your files)
    if not image_name.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
        continue
    
    try:
        # Extract the heart signal from the image
        signal = extract_signal_from_image(image_path)
        
        # Perform frequency domain analysis
        sampling_rate = 1  # Assuming one sample per pixel
        fft_freq, fft_result = frequency_domain_analysis(signal, sampling_rate)
        
        # Detect anomalies in the signal
        z_scores, anomalies = detect_anomalies(signal)
        
        # Calculate the average Z-score
        avg_z_score = np.mean(np.abs(z_scores))
        
        # Store the results
        analysis_results.append({
            'image_name': image_name,
            'z_scores': z_scores,
            'avg_z_score': avg_z_score,
            'fft_freq': fft_freq,
            'fft_result': fft_result
        })
        
        # Plot the signal and highlight anomalies
        plt.figure(figsize=(12, 6))
        plt.plot(signal, label='Heart Signal')
        plt.scatter(anomalies, signal[anomalies], color='red', label='Anomalies')
        plt.title(f'Anomaly Detection in {image_name}')
        plt.xlabel('Index')
        plt.ylabel('Signal Value')
        plt.legend()
        plt.show()
    
    except Exception as e:
        print(f"Error processing {image_name}: {e}")

# Convert the list of dictionaries to a DataFrame
results_df = pd.DataFrame(analysis_results)

# Since z_scores and fft_result are numpy arrays, convert them to strings for CSV storage
results_df['z_scores'] = results_df['z_scores'].apply(lambda zs: ','.join(map(str, zs)))
results_df['fft_result'] = results_df['fft_result'].apply(lambda fr: ','.join(map(str, np.abs(fr))))

# Save results to a CSV file
results_df.to_csv('analysis_results.csv', index=False)

print('Analysis results have been saved to analysis_results.csv')
````
